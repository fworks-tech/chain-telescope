# Market Data Exploration

This notebook imports from the shared `src/` layer to explore market data, price trends, and risk metrics.

Run the setup cell first, then experiment with the data.

```python
# Setup: ensure src/ is importable from the repo root
import sys
from pathlib import Path

repo_root = Path.cwd().resolve()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

print(f"Repo root: {repo_root}")
print("src/ is importable:", (repo_root / "src").exists())
```

## 1. Load a dashboard snapshot

```python
from src.data.dashboard_query import load_dashboard_snapshot
from src.data.mock_market import TIME_WINDOW_DAYS

# Use 'mock' source so no API keys are needed
snap = load_dashboard_snapshot("30D", ["BTC", "ETH", "SOL"], market_source="mock")

print(f"Time window: {snap.time_window}")
print(f"Primary asset: {snap.primary_asset}")
print(f"Data source: {snap.price_source}")
```

## 2. Price trend chart

Plot the price series and moving-average trend using matplotlib.

```python
import matplotlib.pyplot as plt
from src.components.chart_theme import CHART_FONT_COLOR, CHART_GRID_COLOR

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(snap.price_dates, snap.price_values, label=snap.primary_asset, color="#3b5bff")
ax.plot(snap.price_dates, snap.price_values, label="Trend", color="#94a3ff", linestyle="--")
ax.set_title(f"{snap.primary_asset} — Price Trend ({snap.time_window})", color=CHART_FONT_COLOR)
ax.tick_params(colors=CHART_FONT_COLOR)
ax.grid(color=CHART_GRID_COLOR)
ax.legend()
plt.tight_layout()
plt.show()
```

## 3. KPIs and trending report

```python
print("=== KPIs ===")
for kpi in snap.kpis:
    print(f"  {kpi['label']}: {kpi['value']} ({kpi['delta']})")

print("\n=== Trending ===")
display(snap.trending)
```

## 4. Risk graph

```python
fig, ax = plt.subplots(figsize=(6, 3))
ax.barh(snap.risk_labels, snap.risk_scores, color=snap.risk_colors)
ax.set_xlim(0, 100)
ax.set_title("Risk Profile", color=CHART_FONT_COLOR)
ax.tick_params(colors=CHART_FONT_COLOR)
ax.grid(color=CHART_GRID_COLOR, axis="x")
plt.tight_layout()
plt.show()
```

## 5. Alerts and news

```python
print("=== Alerts ===")
for alert in snap.alerts:
    print(f"  - {alert}")

print("\n=== News ===")
for headline in snap.news:
    print(f"  - {headline}")
```
